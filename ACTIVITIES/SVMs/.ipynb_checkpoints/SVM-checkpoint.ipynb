```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Read in the Data

```python
# load iris dataset
iris = datasets.load_iris()

# Create a Python dataframe
iris_df=pd.DataFrame(iris.data)
iris_df.columns=iris.feature_names
iris_df = iris_df.drop(['petal length (cm)','petal width (cm)'], axis=1)
iris_df['species'] = iris.target

iris_df.head()
```

# Plot the data

```python
custom_palette = {0: 'red', 1: 'yellow', 2: 'blue'}
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=iris_df, palette=custom_palette)
```

# Prepare the Data

```python
# Separate features (X) and target variable (y)
X = iris_df.drop('species', axis=1)
y = iris_df['species']
X.head()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Softmax Regression

```python
# Initialize and fit the logistic regression model
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
logreg.fit(X_train, y_train)

# Predictions on the test set
y_pred = logreg.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
```

# Plot the Decision Boundaries

```python
# Get decision boundaries
h = 0.02  # Step size in the mesh
print(X['sepal length (cm)'].min())
x_min, x_max = X['sepal length (cm)'].min() - 1, X['sepal length (cm)'].max() + 1
y_min, y_max = X['petal length (cm)'].min() - 1, X['petal length (cm)'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Get predictions for each point in the mesh
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


# Plot the scatter plot
scatter = sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df, palette=custom_palette)

# Match contour colors with scatter plot colors
contour = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.2)

# Add legend for the scatter plot
scatter.legend()

plt.title('Softmax Regression Decision Boundaries')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

# Fit Support Vector Machine

```python
# Initialize and fit the SVM model
svm_model = SVC(kernel='poly', C=1, degree=3)  # You can adjust the kernel and C parameter as needed
svm_model.fit(X_train, y_train)

# Predictions on the test set
y_pred_svm = svm_model.predict(X_test)

# Evaluate the SVM model
# Evaluate the model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

print(f'Accuracy: {accuracy_svm:.2f}')
print('Confusion Matrix:')
print(conf_matrix_svm)
```

# Plot the Decision Boundaries

```python
# Get predictions for each point in the mesh
Z_svm = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z_svm = Z_svm.reshape(xx.shape)

# Plot the scatter plot
scatter = sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df, palette=custom_palette)

# Match contour colors with scatter plot colors for SVM
contour_svm = plt.contourf(xx, yy, Z_svm, cmap=plt.cm.RdYlBu, alpha=0.2)

# Add legend for the scatter plot
scatter.legend()

plt.title('SVM Decision Boundaries')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
