```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

# Read in the Data

```python
# load  dataset
wine = datasets.load_wine()
print(wine.feature_names)

# Create a Python dataframe
wine_df=pd.DataFrame(wine.data)
wine_df.columns=wine.feature_names
wine_df['quality'] = wine.target
labely = 'od280/od315_of_diluted_wines'
labelx = 'flavanoids'
labeltarget = 'quality'
dataframe = wine_df

dataframe = dataframe[[labely,labelx,labeltarget]]

dataframe.head()
```

# Plot the data

```python
custom_palette = {0: 'red', 1: 'yellow', 2: 'blue'}
sns.scatterplot(y=labely, x=labelx, hue=labeltarget, data=dataframe, palette=custom_palette)
```

# Prepare the Data

```python
# Separate features (X) and target variable (y)
X = dataframe.drop(labeltarget, axis=1)
y = dataframe[labeltarget]
X.head()


# Split the dataset into training, testing, and validation sets
# NOTE: DO NOT TOUCH THE VALIDATION SCRIPT 
# NOTE: DO NOT CHANGE THESE RANDOM_STATE VARIABLES
X_train, X_test_val, y_train, y_test_val = train_test_split(X, y, test_size=0.4, random_state=40)
X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.2, random_state=40)

# Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_scaled = scaler.transform(X)
```

# Softmax Regression

```python
# Initialize and fit the logistic regression model
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
logreg.fit(X_train_scaled, y_train)

# Predictions on the test set
y_pred = logreg.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
```

# Plot the Decision Boundaries

```python
# Get decision boundaries
h = 0.02  # Step size in the mesh
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Get predictions for each point in the mesh
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


# Plot the scatter plot
scatter = sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=y, palette=custom_palette)

# Match contour colors with scatter plot colors
contour = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.2)

# Add legend for the scatter plot
scatter.legend()

plt.title('Softmax Regression Decision Boundaries')
plt.xlabel(labelx)
plt.ylabel(labely)
plt.show()
```

# Fit Support Vector Machine

```python
# Initialize and fit the SVM model
svm_model = SVC(kernel='rbf', C=0.5, degree=3)  # You can adjust the kernel and C parameter as needed
svm_model.fit(X_train_scaled, y_train)

# Predictions on the test set
y_pred_svm = svm_model.predict(X_test_scaled)

# Evaluate the SVM model
# Evaluate the model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

print(f'Accuracy: {accuracy_svm:.2f}')
print('Confusion Matrix:')
print(conf_matrix_svm)
```

# Plot the Decision Boundaries

```python
# Get predictions for each point in the mesh
Z_svm = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z_svm = Z_svm.reshape(xx.shape)

# Plot the scatter plot
scatter = sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=y, palette=custom_palette)

# Match contour colors with scatter plot colors for SVM
contour_svm = plt.contourf(xx, yy, Z_svm, cmap=plt.cm.RdYlBu, alpha=0.2)

# Add legend for the scatter plot
scatter.legend()

plt.title('SVM Decision Boundaries')
plt.xlabel(labelx)
plt.ylabel(labely)
plt.show()
```
