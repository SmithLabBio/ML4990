```python
import pandas
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
```

# Read the Data

As before, we will use pandas and the read_csv function to read the data. We're going to change the species names to zero and one. When you're discussing results, please do it by referring to the actual species names, not zero and 1.

```python
# load dataset
ants = pandas.read_csv('all_measurments2020x09.csv')
ants.head(n=2)

# Change species from names to zero/one.
ants['species'] = ants['species'].replace('zeteki', 0)
ants['species'] = ants['species'].replace('fovouros', 1)
```

# Plot the data

Now, let's plot some of the data to help decide which variables to use. We'll use the library matplotlib for plotting. Specifically, we imported matplotlib.pyplot as plt. This allows us to refer to functions faster. We will plot scatterplots of different pairs of variables, and we will color by the class we are trying to predict.

```python
variable1 = "FLI"
variable2 = "HW"
custom_palette = {0: "red", 1: "blue"}
sns.scatterplot(y=variable1, x=variable2, hue="species", data=ants, palette=custom_palette)
```

# Prepare the Data

Recall that, usually, in machine learning, we divide our dataset into three different portions:
1) the training data, which is used to train the algorithm.
2) the validation data, which is used to evaluate performance and make decisions about how to model the data.
3) the test data, which is used to get a final measure of peformance.

The below code will split your data into training, validation and test datasets.

**Question 1:**

**1a:** What percentage of the dataset are we used for training (train), validation (val) and testing (test) here?

**1b**: Why do we need to divide our data into training, testing, and validation sets?

```python
# Keep only the variables you are going to use!
X = ants[[variable1,variable2,"species"]]

# Separate features (X) and target variable (y)
y = X['species']
X = X.drop('species', axis=1)
X.head()

# Split the dataset into training, testing, and validation sets
# NOTE: DO NOT TOUCH THE TESTING SET 
# NOTE: DO NOT CHANGE THESE RANDOM_STATE VARIABLES
X_train, X_test_val, y_train, y_test_val = train_test_split(X, y, test_size=0.4, random_state=40)
X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5, random_state=40)
```

# Scale the features

Often, in machine learning, we scale features so that they have mean zero and a standard deviation of 1. This causes many algorithms to work better.

**Add**: I didn't comment my code! Add comments. If you aren't sure what these functions do, look it up in the sklearn documentation.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
X_scaled = scaler.transform(X)
```

# Fit a Logistic Regression

First, we will fit a logisitc regression to our data.

**Question 2**: 

**2a:** Write down the model for the logistic regression.

**2b:** Explain where in the model the thetas and theta0 printed below are incorporated.

Optimization: Here 'solver' refers to the optimization technique we are using. We have only discussed gradient ascent and descent, but there are many potential techniques.  All have the same general goal: find the parameters that minimize the cost function or maximize the log likelihood.


```python
# Initialize and fit the logistic regression model
logreg = LogisticRegression(solver='lbfgs', max_iter=10000)
logreg.fit(X_train_scaled, y_train)
print("Thetas/ws:", logreg.coef_, "\nTheta0/b:", logreg.intercept_)
```

# Make predictions using the logistic regression model.

Now, we'll make predictions with our logistic regression model. We will also evaluate the accuracy of our predictions. To do this, we will use accuracy and a confusion matrix.

Accuracy: What percentage of datasets were missclassified?

Confusion Matrix: Breaks this down into classes, and shows how often each label is predicted given the true label.

**Question 3**: Report your accuracy, and confusion matrix. How well does this model perform on the validation data? When discussing, please use species names, not 0 and 1.

```python
# Predictions on the validation set
y_pred_val = logreg.predict(X_val_scaled)

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred_val)

print(f'Accuracy: {accuracy:.2f}')

ConfusionMatrixDisplay.from_estimator(
    logreg, X_val_scaled, y_val)
```

# Function to plot decision boundary.

This is a function for plotting the decision boundaries. 

**Question 4**: Explain how to convert from the inferred coefficients to the slope and intercept of the decision boundary.

```python
def plot_boundary(x, logreg, x1label, x2label):

    # Use the coefficients and intercept inferred when fitting the logreg model to get the slope and intercept for the decision boundary.
    slope = logreg.coef_[0][0]/-logreg.coef_[0][1]
    intercept = logreg.intercept_/-logreg.coef_[0][1]


    # Plot the scatter plot
    scatter = sns.scatterplot(x=x[:,0], y=x[:,1], hue=y, palette=custom_palette)

    # Plot the decision boundary
    x_vals = np.array(scatter.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--', color="black")

    # Add legend for the scatter plot
    scatter.legend()

    plt.title('Logistic Regression Decision Boundaries')
    plt.xlabel(x1label)
    plt.ylabel(x2label)
    plt.show()
```

# Fit Support Vector Machine

In some cases, linear decision boundaries do not fit the data very well. When this happens, we may want to use a SVM instead. Here, we fit a XVM. A few of the important parameters the function takes include:

* kernel: 'polynomial', 'rbf', and more
* C: the regularization parameter (i.e., what is the penalty for misclassification, higher -> a larger penalty)
* degree: The degree of the polynomial when the kernel chosen is polynomial.

```python
svm_model = SVC(kernel='rbf', C=5, degree=3)  # You can adjust the kernel and C parameter as needed
svm_model.fit(X_train_scaled, y_train)
```

# Make predictions on the validation set using the support vector machine

Now, we are ready to make predictions using our support vector machine.

**Change**: Go back to the above code where you fit the SVM. Change the kernel, C, and the degree (when using the polynomial kernel). Fit the SVM with the new settings.

**Question 6**: How does changing the kernel, C, and the degree of the polynomial in the polynomial kernel impact your results? Which model do you think is best, and why?

```python
# Predictions on the validation set
y_pred_val_svm = svm_model.predict(X_val_scaled)

# Evaluate the model
accuracy_svm = accuracy_score(y_val, y_pred_val_svm)

print(f'Accuracy: {accuracy_svm:.2f}')

ConfusionMatrixDisplay.from_estimator(
    svm_model, X_val_scaled, y_val)
```

# Plot the Decision Boundaries

This code will allow you to plot the decision boundaries for the support vector machine. Do not worry if you don't understand the code very well, but make note of how the decision boundaries change when you fit different support vector machines.

**Question 7**: How do the decision boundaries differ between the rbf kernel, the polynomial kernel, and the logistic regression?

```python
# create mesh for calculating decision boundaries
h = 0.02  # Step size in the mesh
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Get predictions for each point in the mesh
Z_svm = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z_svm = Z_svm.reshape(xx.shape)

# Plot the scatter plot
scatter = sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=y, palette=custom_palette)

# Match contour colors with scatter plot colors for SVM
contour_svm = plt.contourf(xx, yy, Z_svm, cmap=plt.cm.RdYlBu, alpha=0.2)

# Add legend for the scatter plot
scatter.legend()

plt.title('SVM Decision Boundaries')
plt.xlabel(variable1)
plt.ylabel(variable2)
plt.show()
```
