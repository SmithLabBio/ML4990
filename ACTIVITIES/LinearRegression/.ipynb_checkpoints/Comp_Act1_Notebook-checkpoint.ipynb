```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
```

# Read and Plot the Data

```python
# read the data
anoles = pd.read_csv('all.csv')
anoles.head(10)

x = anoles['Snout.Vent.Length']
y = anoles["Femur"]
xlabel = "Snout Vent Length (mm)"
ylabel = "Femur Length (mm)"

# plot the data
plt.scatter(x,y)
plt.ylabel(ylabel)
plt.xlabel(xlabel)

```

# Functions for making predictions and plotting predictions

```python
def predictions(x,y,w,b):
    pred = np.array(x)*np.array(w)+np.array(b)
    distances = np.array(y) - pred
    mse = distances**2
    mse = sum(mse)/2/len(pred)
    return(pred, mse)

def plot_predictions(x,y,y_hat,xlabel,ylabel):
    plt.scatter(x,y)
    plt.plot(x, y_hat, color="green")
    [plt.vlines(x=x[i], ymin=y[i], ymax=y_hat[i], color="red", linestyle="--", alpha=0.5) for i in range(len(x))]
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
```

# Try different values of w and b

```python
w_try = 2
b_try = 0
y_hat, mse = predictions(x=x,y=y,w=w_try, b=b_try)
plot_predictions(x=x,y=y, y_hat=y_hat, xlabel=xlabel, ylabel=ylabel)
print("MSE: ", mse)

```

# Functions for Plotting the Cost Function

```python
def cost_function(w,b,x,y):
    pred = np.array(x)*np.array(w)+np.array(b)
    distances = np.array(y) - pred
    mse = distances**2
    mse = sum(mse)/2/len(pred)
    return mse

def calc_cost_grid(X,Y,B=None, W=None, w_min=-10, w_max=10, b_min=-50, b_max=50, w_step=0.1, b_step=0.1):
    if (B==None and W==None):
        w_range = np.arange(w_min,w_max,w_step)
        b_range = np.arange(b_min,b_max,b_step)
        W, B = np.meshgrid(w_range, b_range)
        Z = np.zeros_like(W)
        for i in range(W.shape[0]):
            for j in range(W.shape[1]):
                Z[i, j] = cost_function(W[i, j], B[i, j], X, Y)
    elif B!=None:
        w_range = np.arange(w_min,w_max,w_step)
        W=w_range
        Z = np.zeros_like(w_range)
        for i in range(len(w_range)):
            Z[i] = cost_function(w_range[i], B, X, Y)
    elif W!=None:
        b_range = np.arange(b_min,b_max,b_step)
        B=b_range
        Z = np.zeros_like(b_range)
        for i in range(len(b_range)):
            Z[i] = cost_function(W, b_range[i], X, Y)
    return Z,W,B
```

# Plot the Cost Function

```python
Z,W,B = calc_cost_grid(x, y,b_min=-50, b_max=50, w_min=-5, w_max=5, b_step=1, w_step=0.1)

# surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(W, B, Z, cmap='viridis')
ax.set_xlabel('W-axis')
ax.set_ylabel('B-axis')
ax.set_zlabel('Cost')
ax.set_title('3D Surface Plot of Cost Function')
fig.colorbar(surf)
ax.view_init(elev=20, azim=10)  # Change the elevation and azimuth angles
plt.show()

# contour plot
plt.contour(W, B, Z, levels=800, linewidths=1, alpha=0.7, cmap='viridis')
plt.xlabel('W-axis')
plt.ylabel('B-axis')
plt.title('Contour Plot of Cost Function')
plt.colorbar()
plt.show()
```

# Plot 2D Cost function for w

```python
Z_fixedB,W_fixedB,B_fixedB = calc_cost_grid(x, y,B=3)
Z_fixedW,W_fixedW,B_fixedW = calc_cost_grid(x, y,W=-.04)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # Adjust figsize as needed
axes[0].plot(W_fixedB, Z_fixedB, label='B=3', color='purple')
axes[0].set_title('Fixed B')
axes[0].legend()
axes[0].set_ylabel("Z")
axes[0].set_xlabel("W")
axes[1].plot(B_fixedW, Z_fixedW, label='W=-.04', color='orange')
axes[1].set_title('Fixed W')
axes[1].legend()
axes[1].set_ylabel("Z")
axes[1].set_xlabel("B")
plt.tight_layout()
plt.show()
```

# Gradient Descent

```python
def fit(X, y, iterations, b_init=0, w_init=0, learning_rate=0.005):
    mse = []
    b_states = []
    w_states = []
    b = b_init
    w = w_init
    n = X.shape[0]
    b_states.append(b)
    w_states.append(w)
    for j in range(iterations):
        mse.append(np.sum((np.array(y) - np.array(w)*np.array(X) + np.array(b))**2)/2/n)
        b_gradient = -2 * np.sum(y - w*X + b) / n
        w_gradient = -2 * np.sum(X*(y - (w*X + b))) / n
        b = b + (learning_rate * b_gradient)
        w = w - (learning_rate * w_gradient)
        b_states.append(b)
        w_states.append(w)
    mse.append(np.sum((y - w*X + b)**2)/2/n)
    return(w,b, mse, w_states, b_states)
```

```python
iter = 10000
lr=0.01
fit_w, fit_b, mse_fit, w_states, b_states = fit(np.array(x), np.array(y), iterations=iter, learning_rate=lr, w_init=-4, b_init=-10)
print(fit_w, fit_b)

y_hat, mse = predictions(x, y,fit_w,fit_b)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # Adjust figsize as needed

axes[0].scatter(x, y)
axes[0].plot(x, y_hat, color="green")
[axes[0].vlines(x=x[i], ymin=y[i], ymax=y_hat[i], color="red", linestyle="--", alpha=0.5) for i in range(len(y_hat))]
axes[0].set_ylabel(ylabel)
axes[0].set_xlabel(xlabel)
    
axes[1].plot(range(iter+1), mse_fit, color="blue")
axes[1].set_ylabel('MSE')
axes[1].set_xlabel("Iteration")
```
